```python
"""
04 - Full-Sample Significance Test
----------------------------------
Tests whether mean CAR and CSAR are significantly
different from zero for each event window.

Test method
For mean CAR  : Cross-sectional standard deviation test
For mean CSAR : Standardized cross-sectional test

Data source : clean_event_data.csv
"""

import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv("clean_event_data.csv")

windows = {
    "(-1,1)": (-1, 1),
    "(-5,5)": (-5, 5),
    "(-10,10)": (-10, 10)
}

for name, (start, end) in windows.items():
    
    temp = df[(df["Tau"] >= start) & (df["Tau"] <= end)]

    #CAR test
    car = temp.groupby("firm")["AR"].sum()
    N = len(car)
    mean_car = car.mean()
    std_car = car.std(ddof=1)

    t_car = mean_car / (std_car / np.sqrt(N)) # ~N(0,1)
    p_car = 2 * stats.norm.sf(abs(t_car))
    
    #CSAR test
    csar = temp.groupby("firm")["SAR"].sum()
    mean_csar = csar.mean()
    std_csar = csar.std(ddof=1)

    t_csar = mean_csar / (std_csar / np.sqrt(N)) # ~N(0,1)
    p_csar = 2 * stats.norm.sf(abs(t_csar))

    print(f"\nWindow {name}")
    print(f"N = {N}")
    print(f"Mean CAR = {mean_car:.4f} | t = {t_car:.4f} | p = {p_car:.4f}")
    print(f"Mean CSAR = {mean_csar:.4f} | t = {t_csar:.4f} | p = {p_csar:.4f}")
```
