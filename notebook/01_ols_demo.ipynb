```python
"""
01 - OLS Market Model Demo (Foxconn)
------------------------------------
Demonstrates the AR/SAR calculation procedure using Foxconn.
Actual AR/SAR for all firms were computed via TEJ; 
this notebook serves as a methodology verification.

- Estimation Window: [-240, -41]
- Event Windows    : (-1, +1), (-5, +5), (-10, +10)
- Model            : Market Model using OLS
"""

import pandas as pd
import statsmodels.api as sm
import numpy as np

df = pd.read_excel("ols_data.xlsx")
df.head()

#Estimate alpha & beta
estimation = df[(df["Tau"] >= -240) & (df["Tau"] <= -41)]
X = estimation["Market_r"]
X = sm.add_constant(X)
y = estimation["Return"]

model = sm.OLS(y, X).fit()
alpha = model.params["const"]
beta = model.params["Market_r"]
print(alpha, beta)

sigma = np.std(model.resid, ddof=1)
print("sigma:", sigma)

#Calculate expected returns and AR/SAR
df["expected_return"] = alpha + beta * df["Market_r"]
df["AR"] = df["Return"] - df["expected_return"]

df["SAR"] = df["AR"] / sigma

#Aggregate CAR/CSAR
CAR_1 = df[(df["Tau"] >= -1) & (df["Tau"] <= 1)]["AR"].sum()
CAR_5 = df[(df["Tau"] >= -5) & (df["Tau"] <= 5)]["AR"].sum()
CAR_10 = df[(df["Tau"] >= -10) & (df["Tau"] <= 10)]["AR"].sum()
print(CAR_1, CAR_5, CAR_10)

CSAR_1 = df[(df["Tau"] >= -1) & (df["Tau"] <= 1)]["SAR"].sum()
CSAR_5 = df[(df["Tau"] >= -5) & (df["Tau"] <= 5)]["SAR"].sum()
CSAR_10 = df[(df["Tau"] >= -10) & (df["Tau"] <= 10)]["SAR"].sum()
print(CSAR_1, CSAR_5, CSAR_10)
```
