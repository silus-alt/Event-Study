```python
"""
02 - Data Cleaning
------------------
Reshapes wide-format data (from TEJ) into long format.

Input  : AR_tej.xlsx
Output : clean_event_data.csv
"""
import pandas as pd
ar = pd.read_excel("AR_tej.xlsx", sheet_name=0)
sar = pd.read_excel("AR_tej.xlsx", sheet_name=1)

ar_long = ar.melt(id_vars="Tau", var_name="firm", value_name="AR")
sar_long = sar.melt(id_vars="Tau", var_name="firm", value_name="SAR")
df = pd.merge(ar_long, sar_long, on=["Tau", "firm"])

df.head()
df.to_csv("clean_event_data.csv", index=False)
```
