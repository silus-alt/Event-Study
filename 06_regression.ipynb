```python
"""
06 - Cross-Sectional Regression Analysis
----------------------------------------
OLS regression of CAR on alliance-level and firm-level characteristics.
DV : CAR
IV : LEAD, TECH, RD, BM, TECHxRD

Data source : firm_fund.xlsx
"""

import pandas as pd
import statsmodels.api as sm
 
df = pd.read_excel("firm_fund.xlsx")

#Centering for interaction
df["RD_c"] = df["RD"] - df["RD"].mean()
df["TECH_RD"] = df["TECH"] * df["RD_c"]

X = sm.add_constant(df[["LEAD", "TECH", "RD", "BM", "TECH_RD"]])

windows = {
    "(-1,1)": "CAR -1~+1",
    "(-5,5)": "CAR -5~+5",
    "(-10,10)": "CAR -10~+10",
}

for name, col in windows.items():
    y = df[col]
    model = sm.OLS(y, X).fit()

    print(f"Window {name}")
    print(f"{'='*41}")
    print(f"{'Variable':<12} {'Coef':>10} {'t':>8} {'p':>8}")
    print(f"{'-'*41}")

    for var in model.params.index:
        coef = model.params[var]
        t_val = model.tvalues[var]
        p_val = model.pvalues[var]
        star = "***" if p_val < 0.01 else "**" if p_val < 0.05 else "*" if p_val < 0.1 else ""

        print(f"{var:<12} {coef:>10.2f} {t_val:>8.2f} {p_val:>8.4f} {star}")

    print(f"{'='*41}")
    print(f"N = {int(model.nobs)}")
    print(f"R^2 = {model.rsquared:.1%}")
    print(f"Adj R^2 = {model.rsquared_adj:.1%}")
    print(f"{'='*41}\n")
```
