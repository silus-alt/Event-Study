```python
"""
03 - Line Chart
---------------
Visualize average AR/SAR across event days.

Data source : clean_event_data.csv
"""
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

df = pd.read_csv("clean_event_data.csv")

aar = df.groupby("Tau")["AR"].mean().sort_index()
asar = df.groupby("Tau")["SAR"].mean().sort_index()

rcParams['font.family'] = 'DejaVu Serif'
plt.figure(figsize=(10,5))

plt.plot(
    aar.index, 
    aar.values, 
    color="black", 
    linewidth=1.8, 
    label="Average AR (%)"
)
plt.plot(
    asar.index, 
    asar.values, 
    color="black", 
    linestyle="--", 
    linewidth=1.8, 
    label="Average SAR"
)

plt.axvline(x=0, color="gray", linewidth=0.9)
plt.xlabel("Event Day", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.xticks(aar.index, fontsize=12)
plt.yticks(fontsize=12)

ylim = max(abs(min(aar.values + asar.values)), abs(max(aar.values + asar.values)))
plt.ylim(-ylim, ylim)

plt.grid(True, linewidth=0.6)
plt.legend(fontsize=13)

plt.tight_layout()
plt.show()
```
